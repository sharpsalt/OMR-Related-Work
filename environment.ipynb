```python
import cv2
import matplotlib.pyplot as plt

image = cv2.imread("test.jpg", cv2.IMREAD_GRAYSCALE)

# Apply Gaussian Blur to reduce noise (optional but improves Otsu)
blur = cv2.GaussianBlur(image, (5, 5), 0)

# Apply Otsu's thresholding
_, binary = cv2.threshold(blur, 0, 255, cv2.THRESH_BINARY + cv2.THRESH_OTSU)

# Save the result
cv2.imwrite("binary_output.jpg", binary)

# Display the result using matplotlib
plt.figure(figsize=(10,5))
plt.subplot(1,2,1)
plt.title("Original Grayscale")
plt.imshow(image, cmap='gray')

plt.subplot(1,2,2)
plt.title("Black & White (Otsu)")
plt.imshow(binary, cmap='gray')

plt.show()
```

```python


# Define crop coordinates
x1, y1 = 160, 98
x2, y2 = 194, 132

# Crop the image using slicing [y1:y2, x1:x2]
cropped = image[y1:y2, x1:x2]

# Save the cropped image
cv2.imwrite("cropped_output.jpg", cropped)

# (Optional) Display the result
import matplotlib.pyplot as plt

plt.imshow(cv2.cvtColor(cropped, cv2.COLOR_BGR2RGB))
# plt.title("Cropped Region")
plt.axis('off')
plt.show()
```

```python
print(cropped.shape)
```

```python

import numpy as np
# Load the image
img = cropped

# Resize it to exactly 34x34 (just in case), without distortion
img = cv2.resize(img, (34, 34), interpolation=cv2.INTER_NEAREST)

# Convert to color (if not already)
if len(img.shape) == 2 or img.shape[2] == 1:
    img = cv2.cvtColor(img, cv2.COLOR_GRAY2BGR)

# Simulate an OMR-like blue border
# We'll draw a blue circle around the center
center = (17, 17)
radius = 16
thickness = 2
cv2.circle(img, center, radius, (255, 0, 0), thickness)  # Blue circle

# Optionally, fill a light-gray center to simulate marked bubble
# Uncomment to simulate a filled bubble:
# cv2.circle(img, center, radius - 4, (200, 200, 200), -1)

# Save or display
cv2.imwrite("colored_omr_bubble.jpg", img)

# Display
plt.imshow(cv2.cvtColor(img, cv2.COLOR_BGR2RGB))
plt.title("Colored OMR Bubble (34x34)")
plt.axis('off')
plt.show()
```
